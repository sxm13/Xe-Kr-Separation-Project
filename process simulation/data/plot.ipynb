```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
import seaborn as sns
```

# import data for all

```python
data = pd.read_excel("IVSAresults.xlsx",sheet_name="all")
```

# get data: Xe purity, Selectivity, Henry selectivity and Working capacity

```python
Xe = data["Xe_purity"]*100

S1 = data["Selectivity_1bar"]
S2 = data["Selectivity_0.1bar"]

SH = data["Xe_henry"]/data["Kr_henry"]

WC1 = data["WC_Xe"]
WC2 = data["WC_Kr"]
```

# calculate SRCC

```python
SRCC_1 ,p=spearmanr(S1,Xe)
SRCC_2 ,p=spearmanr(S2,Xe)
SRCC_3 ,p=spearmanr(SH,Xe)
SRCC_4 ,p=spearmanr(WC1,Xe)
SRCC_5 ,p=spearmanr(WC2,Xe)
print(SRCC_1,SRCC_2,SRCC_3,SRCC_4,SRCC_5)
```

# parity plot

```python
fig = plt.figure(figsize=(18,5),dpi=600)

ax = fig.add_subplot(131)
bx = fig.add_subplot(132)
cx = fig.add_subplot(133)


ax1 = ax.scatter(S1,Xe,s=10,c="b")
ax2 = ax.scatter(S2,Xe,s=10,c="r")

ax3 = ax.scatter(100,100,s=40,c="b",label="1 bar")
ax4 = ax.scatter(100,100,s=40,c="r",label="0.1 bar")


ax.set_xlim(0,20)
ax.set_ylim(20,80)
ax.set_ylabel(r"$\mathregular {Purity_{max,Xe,ML}}$ / %",fontdict={'family':'Calibri','size':22})
x=[20,30,40,50,60,70]
ax.set_yticks(x)
ax.spines['top'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
ax.minorticks_off()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.text(0.05,0.9,"a",fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=ax.transAxes)
ax.legend(loc=(0.6,0.02), prop={'family':'Calibri','size':20}, frameon=True)
ax.set_xlabel(r'Xe/Kr IAST Selectivity',fontdict={'family':'Calibri','size':22})

bx1 = bx.scatter(SH,Xe,s=10,c="b")
# bx.scatter(SH1,Xe1,s=10,c=data1["Density"])
# bx.scatter(SH1,Xe1,s=10,c=data1["PLD"])
# bx.scatter(SH1,Xe1,s=10,c=data1["VF"])

bx.set_xlim(0,20)
bx.set_ylim(20,80)
bx.set_yticks([])
bx.spines['top'].set_linewidth(1.5)
bx.spines['bottom'].set_linewidth(1.5)
bx.spines['left'].set_linewidth(1.5)
bx.spines['right'].set_linewidth(1.5)
bx.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
bx.minorticks_off()
bx.text(0.05,0.9,"b",fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=bx.transAxes)
bx.spines['right'].set_visible(False)
bx.spines['top'].set_visible(False)
bx.set_xlabel(r'Xe/Kr Henry Selectivity',fontdict={'family':'Calibri','size':22})

cx1 = cx.scatter(WC1,Xe,s=10,c="b")
cx2 = cx.scatter(WC2,Xe,s=10,c="r")

cx3 = cx.scatter(100,100,s=40,c="b",label="Xe")
cx4 = cx.scatter(100,100,s=40,c="r",label="Kr")

cx.set_xlim(0,4)
cx.set_ylim(20,80)
cx.set_yticks([])
cx.spines['top'].set_linewidth(1.5)
cx.spines['bottom'].set_linewidth(1.5)
cx.spines['left'].set_linewidth(1.5)
cx.spines['right'].set_linewidth(1.5)
cx.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
cx.minorticks_off()
cx.text(0.05,0.9,"c",fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=cx.transAxes)
cx.spines['right'].set_visible(False)
cx.spines['top'].set_visible(False)
cx.legend(loc=(0.7,0.02), prop={'family':'Calibri','size':20}, frameon=True)
cx.set_xlabel(r'Working Capacity / mmol·g$^{-1}$',fontdict={'family':'Calibri','size':22})

plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.1, 
                    hspace=0.1)
plt.show()
```
