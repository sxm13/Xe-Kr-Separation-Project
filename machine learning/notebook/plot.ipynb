```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
data1 = pd.read_excel('compareTrueandPre.xlsx',sheet_name="Train_Xe")
data2 = pd.read_excel('compareTrueandPre.xlsx',sheet_name="Test_Xe")

data3 = pd.read_excel('compareTrueandPre.xlsx',sheet_name="Train_Kr")
data4 = pd.read_excel('compareTrueandPre.xlsx',sheet_name="Test_Kr")
```

```python
r2_Xe_train = r2_score(data1["Xe_M"],data1["Xe_M_pre"])
r2_Kr_train = r2_score(data3["Kr_M"],data3["Kr_M_pre"])

r2_Xe_test = r2_score(data2["Xe_M"],data2["Xe_M_pre"])
r2_Kr_test = r2_score(data4["Kr_M"],data4["Kr_M_pre"])

mae_Xe_train = mean_absolute_error(data1["Xe_M"],data1["Xe_M_pre"])
mae_Kr_train = mean_absolute_error(data3["Kr_M"],data3["Kr_M_pre"])

mae_Xe_test = mean_absolute_error(data2["Xe_M"],data2["Xe_M_pre"])
mae_Kr_test = mean_absolute_error(data4["Kr_M"],data4["Kr_M_pre"])

rmse_Xe_train = mean_squared_error(data1["Xe_M"],data1["Xe_M_pre"],squared=False)
rmse_Kr_train = mean_squared_error(data3["Kr_M"],data3["Kr_M_pre"],squared=False)

rmse_Xe_test = mean_squared_error(data2["Xe_M"],data2["Xe_M_pre"],squared=False)
rmse_Kr_test = mean_squared_error(data4["Kr_M"],data4["Kr_M_pre"],squared=False)
```

```python
fig = plt.figure(figsize=(13,12),dpi=600)

ax = fig.add_subplot(221)
bx = fig.add_subplot(222)
cx = fig.add_subplot(223)
dx = fig.add_subplot(224)

ax.plot([-10,1000],[-10,1000],'--',color='0.5',zorder=0)

ax.set_xscale('log')
ax.set_yscale('log')

X_mk1,Y_mk1 = data1["Xe_M"],data1["Xe_M_pre"]
ax1=ax.hexbin(X_mk1,Y_mk1, gridsize=(40,20), vmin=0,vmax=55,
              xscale='log',yscale='log',mincnt=0.1,edgecolors='None',cmap='viridis_r')

ax.set_xlim(0.1,100)
ax.set_ylim(0.1,100)

ax.set_xlabel(r'$\mathregular{M_{Fit,Xe}}$ / mmolg$^{-1}$',fontdict={'family':'Calibri','size':22})
ax.set_ylabel(r"$\mathregular{M_{ML,Xe,Train}}$ / mmolg$^{-1}$",fontdict={'family':'Calibri','size':22})

x=[1,10,100]

ax.set_xticks(x)
ax.set_yticks(x)
ax.spines['top'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)
ax.spines['right'].set_linewidth(1.5)
ax.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
ax.minorticks_off()

ax.text(0.05,0.9,"a",
        fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=ax.transAxes)
ax.text(0.55,0.22,"R$^{2}$ = "+format(r2_Xe_train,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=ax.transAxes)
ax.text(0.55,0.14,"MAE = "+format(mae_Xe_train,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=ax.transAxes)
ax.text(0.55,0.06,"RMSE = "+format(rmse_Xe_train,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=ax.transAxes)


bx.plot([-10,1000],[-10,1000],'--',color='0.5',zorder=0)

bx.set_xscale('log')
bx.set_yscale('log')

X_mk2,Y_mk2 = data2["Xe_M"],data2["Xe_M_pre"]
bx1=bx.hexbin(X_mk2,Y_mk2, gridsize=(30,15), vmin=0,vmax=55,
              xscale='log',yscale='log', mincnt=0.1,edgecolors='None',cmap='viridis_r')

bx.set_xlim(0.1,100)
bx.set_ylim(0.1,100)

bx.set_xlabel(r'$\mathregular{M_{Fit,Xe}}$ / mmolg$^{-1}$',fontdict={'family':'Calibri','size':22})
bx.set_ylabel(r"$\mathregular{M_{ML,Xe,Test}}$ / mmolg$^{-1}$",fontdict={'family':'Calibri','size':22})

bx.set_xticks(x)
bx.set_yticks(x)
bx.spines['top'].set_linewidth(1.5)
bx.spines['bottom'].set_linewidth(1.5)
bx.spines['left'].set_linewidth(1.5)
bx.spines['right'].set_linewidth(1.5)
bx.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
bx.minorticks_off()

bx.text(0.05,0.9,"b",fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=bx.transAxes)
bx.text(0.55,0.22,"R$^{2}$ = "+format(r2_Xe_test,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=bx.transAxes)
bx.text(0.55,0.14,"MAE = "+format(mae_Xe_test,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=bx.transAxes)
bx.text(0.55,0.06,"RMSE = "+format(rmse_Xe_test,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=bx.transAxes)


cx.plot([-10,1000],[-10,1000],'--',color='0.5',zorder=0)

cx.set_xscale('log')
cx.set_yscale('log')

X_mk3,Y_mk3 = data3["Kr_M"],data3["Kr_M_pre"]
cx1=cx.hexbin(X_mk3,Y_mk3, gridsize=(30,15), vmin=0,vmax=55,
              xscale='log',yscale='log', mincnt=0.1,edgecolors='None',cmap='viridis_r')

cx.set_xlim(0.1,100)
cx.set_ylim(0.1,100)

cx.set_xlabel(r'$\mathregular{M_{Fit,Kr}}$ / mmolg$^{-1}$',fontdict={'family':'Calibri','size':22})
cx.set_ylabel(r"$\mathregular{M_{ML,Kr,Train}}$ / mmolg$^{-1}$",fontdict={'family':'Calibri','size':22})

cx.set_xticks(x)
cx.set_yticks(x)
cx.spines['top'].set_linewidth(1.5)
cx.spines['bottom'].set_linewidth(1.5)
cx.spines['left'].set_linewidth(1.5)
cx.spines['right'].set_linewidth(1.5)
cx.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
cx.minorticks_off()

cx.text(0.05,0.9,"c",fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=cx.transAxes)
cx.text(0.55,0.22,"R$^{2}$ = "+format(r2_Kr_train,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=cx.transAxes)
cx.text(0.55,0.14,"MAE = "+format(mae_Kr_train,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=cx.transAxes)
cx.text(0.55,0.06,"RMSE = "+format(rmse_Kr_train,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=cx.transAxes)


dx.plot([-10,1000],[-10,1000],'--',color='0.5',zorder=0)

dx.set_xscale('log')
dx.set_yscale('log')

X_mk4,Y_mk4 = data4["Kr_M"],data4["Kr_M_pre"]
dx1=dx.hexbin(X_mk4,Y_mk4, gridsize=(30,15), vmin=0,vmax=55,
              xscale='log',yscale='log', mincnt=0.1,edgecolors='None',cmap='viridis_r')

dx.set_xlim(0.1,100)
dx.set_ylim(0.1,100)

dx.set_xlabel(r'$\mathregular{M_{Fit,Kr}}$ / mmolg$^{-1}$',fontdict={'family':'Calibri','size':22})
dx.set_ylabel(r"$\mathregular{M_{ML,Kr,Test}}$ / mmolg$^{-1}$",fontdict={'family':'Calibri','size':22})

dx.set_xticks(x)
dx.set_yticks(x)
dx.spines['top'].set_linewidth(1.5)
dx.spines['bottom'].set_linewidth(1.5)
dx.spines['left'].set_linewidth(1.5)
dx.spines['right'].set_linewidth(1.5)
dx.tick_params(axis='both', which='both', direction='out',width=1.5,labelsize=16)
dx.minorticks_off()

dx.text(0.05,0.9,"d",fontdict={'family':'Calibri','size':24,'fontweight':'bold'},transform=dx.transAxes)
dx.text(0.55,0.22,"R$^{2}$ = "+format(r2_Kr_test,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=dx.transAxes)
dx.text(0.55,0.14,"MAE = "+format(mae_Kr_test,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=dx.transAxes)
dx.text(0.55,0.06,"RMSE = "+format(rmse_Kr_test,'.3f'),
        fontdict={'family':'Calibri','size':20},transform=dx.transAxes)



bx.text(1.33,-0.2,"Counts",fontdict={'family':'Calibri','size':24},ha='center',va='center',rotation=90,transform=bx.transAxes)
cbar_ax = fig.add_axes([0.93, 0.25, 0.015, 0.5])
fig.colorbar(ax1, ax=[ax,bx], cax=cbar_ax,ticks=[0,10,20,30,40,50])
cbar_ax.tick_params(labelsize=16)
# cbar_ax.set_ticks()

plt.subplots_adjust(left=0.125,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.5, 
                    hspace=0.35)
plt.show()
```
